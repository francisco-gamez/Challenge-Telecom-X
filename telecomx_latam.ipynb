#ðŸ“Œ ExtracciÃ³n

```python
import pandas as pd
```

```python
df = pd.read_json('https://raw.githubusercontent.com/francisco-gamez/Challenge-Telecom-X/refs/heads/main/TelecomX_Data.json')
df
```

```python
df_customer = df['customer'].apply(pd.Series)
df_phone = df['phone'].apply(pd.Series)
df_internet = df['internet'].apply(pd.Series)
df_account = df['account'].apply(pd.Series)
```

```python
df = pd.concat([df.drop(columns=['customer', 'phone', 'internet','account']),
                   df_customer, df_phone, df_internet,df_account], axis=1)
df
```

```python
df_Charges = df['Charges'].apply(pd.Series)
df = pd.concat([df.drop(columns=['Charges']), df_Charges], axis=1)
df
```

```python
df.info()
```

#ðŸ”§ TransformaciÃ³n

```python
# Cambiando el nombre de las columnas conforme al diccionario

traduccion_columnas = {
    'customerID': 'ID_cliente',
    'Churn': 'Abandono',
    'gender': 'GÃ©nero',
    'SeniorCitizen': 'Adulto_mayor',
    'Partner': 'Pareja',
    'Dependents': 'Dependientes',
    'tenure': 'Meses_contratados',
    'PhoneService': 'Servicio_telefÃ³nico',
    'MultipleLines': 'MÃºltiples_lÃ­neas',
    'InternetService': 'Servicio_Internet',
    'OnlineSecurity': 'Seguridad_en_lÃ­nea',
    'OnlineBackup': 'Respaldo_en_lÃ­nea',
    'DeviceProtection': 'ProtecciÃ³n_de_dispositivo',
    'TechSupport': 'Soporte_tÃ©cnico',
    'StreamingTV': 'TelevisiÃ³n_por_cable',
    'StreamingMovies': 'Streaming_de_pelÃ­culas',
    'Contract': 'Tipo_contrato',
    'PaperlessBilling': 'Factura_electrÃ³nica',
    'PaymentMethod': 'MÃ©todo_de_pago',
    'Monthly': 'Cargos_mensuales',
    'Total': 'Cargos_totales'
}
df.rename(columns=traduccion_columnas, inplace=True)
df.columns
```

```python
# Verifique si hay valores Ãºnicos en cada columna

for col in df.columns:
    print(f"Columna: {col}")
    print(f"Valores Ãºnicos: {df[col].nunique(dropna=False)}")
    print("Valores Ãºnicos:", df[col].unique()[:5])
    print("-" * 50)
```

```python
# Verificamos cuÃ¡ntas filas duplicadas existen
df.duplicated().sum()
```

```python
# Creamos una tabla resumen de nulos
resumen_nulos = pd.DataFrame({
    'Columna': df.columns,
    'Nulos': df.isnull().sum(),
    'Tipo de dato': df.dtypes
}).reset_index(drop=True)

resumen_nulos
```

```python
# Resumen de vacÃ­os
resumen_nulos = pd.DataFrame({
    'Columna': df.columns,
    'Nulos (NaN)': df.isnull().sum().values,
    'Espacios en blanco': (df.applymap(lambda x: isinstance(x, str) and x.strip() == '')).sum().values,
    'Total vacÃ­os': (df.isnull().sum() + df.applymap(lambda x: isinstance(x, str) and x.strip() == '')).sum().values,
})

# Mostrar solo las columnas que tienen vacÃ­os
resumen_vacios = resumen_nulos[resumen_nulos['Total vacÃ­os'] > 0]

resumen_vacios
```

```python
# Eliminar filas donde Abandono o Cargos_totales sean cadenas vacÃ­as ('')
df = df[~(
    (df['Abandono'].str.strip() == '') |
    (df['Cargos_totales'].str.strip() == '')
)]
df.info()
```

```python
# Crear la columna 'Cuentas_Diarias' dividiendo cargos mensuales entre 30
df['Cuentas_Diarias'] = df['Cargos_mensuales'] / 30
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


# ConfiguraciÃ³n de estilo
sns.set(style="white")
plt.rcParams["figure.figsize"] = (10, 6)

# ConfiguraciÃ³n de colores
color_no = "#4098cf"
color_yes = "#c9521a"
```

```python
df_filtrado = df[df["Abandono"].isin(["Yes", "No"])]

# Histograma
sns.histplot(df, x='Meses_contratados',
             hue='Abandono',
             multiple='dodge',
             bins=30,
             palette={'Yes':color_yes, 'No':color_no},
             kde=False,
             alpha=0.6,)

plt.title('ComparaciÃ³n de Meses Contratados: Abandono vs Permanencia', fontsize=14)
plt.xlabel('Meses Contratados')
plt.ylabel('NÃºmero de Clientes')
plt.legend(title='Estado del Cliente',labels=['Abandona','Permanece'])
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# GrÃ¡fico de densidad (KDE)
sns.kdeplot(data=df[df['Abandono'] == 'Yes'],
            x='Meses_contratados',
            label='Abandona',
            fill=True,
            common_norm=False,
            alpha=0.5, color=color_yes, linewidth=2)

sns.kdeplot(data=df[df['Abandono'] == 'No'],
            x='Meses_contratados',
            label='Permanece',
            fill=True,
            common_norm=False, alpha=0.5, color=color_no, linewidth=2)

plt.title('Densidad de Meses Contratados: Abandono vs Permanencia', fontsize=14)
plt.xlabel('Meses Contratados')
plt.ylabel('Densidad')
plt.legend(title='Estado del Cliente')
plt.grid(True)
plt.tight_layout()
```

## ðŸ”¹ IntroducciÃ³n

La retenciÃ³n de clientes es un desafÃ­o constante en el sector de telecomunicaciones, especialmente cuando los usuarios tienen contratos flexibles y mÃºltiples opciones en el mercado. Este anÃ¡lisis se enfoca en identificar los factores que influyen en la **evasiÃ³n de clientes (churn)** utilizando un conjunto de datos de clientes de la empresa **TelecomX**.

El objetivo principal es comprender los perfiles de los clientes que abandonan la compaÃ±Ã­a, los servicios que mÃ¡s se asocian al abandono y cÃ³mo ciertos patrones de uso pueden predecir el riesgo de churn. Este conocimiento permitirÃ¡ generar estrategias que mejoren la retenciÃ³n y optimicen la experiencia del cliente.

---

## ðŸ”¹ Limpieza y Tratamiento de Datos

Los datos fueron importados desde un archivo `.json` mediante una URL pÃºblica de GitHub. Posteriormente, se realizaron las siguientes acciones de limpieza y transformaciÃ³n:

- EliminaciÃ³n de filas con valores nulos.
- ConversiÃ³n de variables numÃ©ricas incorrectamente tipadas (`Cargos_totales`) a tipo float.
- NormalizaciÃ³n de etiquetas (`Yes`/`No`, `No internet service`).
- ConversiÃ³n de valores binarios como `Adulto_mayor` de 0/1 a `SÃ­`/`No`.
- CreaciÃ³n de nuevas variables como:
  - `Cuentas_Diarias`: costo diario del servicio (calculado a partir del cargo mensual).
  - `Servicios_Activos`: nÃºmero de servicios digitales contratados por cliente.
  - `Abandono_Num`: variable binaria para anÃ¡lisis de correlaciones.

## ðŸ”¹ Conclusiones e Insights

- El **perfil de cliente con mayor probabilidad de abandono** es alguien con contrato *mes a mes*, menos de 10 meses en la compaÃ±Ã­a, servicios de fibra Ã³ptica, y mÃºltiples servicios digitales (aunque esto Ãºltimo es paradÃ³jico, indicando frustraciÃ³n).
- Los clientes que **pagan mÃ¡s por dÃ­a** tienden a abandonar con mayor frecuencia, lo cual sugiere una **sensibilidad al precio diario**, mÃ¡s que al mensual.
- En contraste, los clientes que **contratan mÃ¡s servicios activos** tienen menor tasa de abandono, lo que indica que **el valor percibido** puede fidelizar si se entrega correctamente.

## ðŸ”¹ Recomendaciones

Con base en los anÃ¡lisis realizados, se sugieren las siguientes acciones estratÃ©gicas:

1. **Fidelizar en los primeros 6 meses**: implementar programas de bienvenida, atenciÃ³n personalizada o incentivos para clientes en contratos flexibles.

3. **Optimizar la experiencia con servicios adicionales**: mejorar usabilidad, comunicaciÃ³n del valor agregado y soporte.
4. **Revisar el proceso de pago electrÃ³nico**: detectar fricciones o errores que generen desconfianza.
5. **Estrategias de upselling inteligentes**: promover paquetes de servicios adicionales con beneficios integrados que aumenten la retenciÃ³n.
6. **Ofrecer descuentos o mejoras a clientes de alto gasto diario**, ya que son mÃ¡s propensos a dejar la compaÃ±Ã­a si no perciben suficiente valor.
