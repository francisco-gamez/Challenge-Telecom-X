#📌 Extracción

```python
import pandas as pd
```

```python
df = pd.read_json('https://raw.githubusercontent.com/francisco-gamez/Challenge-Telecom-X/refs/heads/main/TelecomX_Data.json')
df
```

```python
df_customer = df['customer'].apply(pd.Series)
df_phone = df['phone'].apply(pd.Series)
df_internet = df['internet'].apply(pd.Series)
df_account = df['account'].apply(pd.Series)
```

```python
df = pd.concat([df.drop(columns=['customer', 'phone', 'internet','account']),
                   df_customer, df_phone, df_internet,df_account], axis=1)
df
```

```python
df_Charges = df['Charges'].apply(pd.Series)
df = pd.concat([df.drop(columns=['Charges']), df_Charges], axis=1)
df
```

```python
df.info()
```

#🔧 Transformación

```python
# Cambiando el nombre de las columnas conforme al diccionario

traduccion_columnas = {
    'customerID': 'ID_cliente',
    'Churn': 'Abandono',
    'gender': 'Género',
    'SeniorCitizen': 'Adulto_mayor',
    'Partner': 'Pareja',
    'Dependents': 'Dependientes',
    'tenure': 'Meses_contratados',
    'PhoneService': 'Servicio_telefónico',
    'MultipleLines': 'Múltiples_líneas',
    'InternetService': 'Servicio_Internet',
    'OnlineSecurity': 'Seguridad_en_línea',
    'OnlineBackup': 'Respaldo_en_línea',
    'DeviceProtection': 'Protección_de_dispositivo',
    'TechSupport': 'Soporte_técnico',
    'StreamingTV': 'Televisión_por_cable',
    'StreamingMovies': 'Streaming_de_películas',
    'Contract': 'Tipo_contrato',
    'PaperlessBilling': 'Factura_electrónica',
    'PaymentMethod': 'Método_de_pago',
    'Monthly': 'Cargos_mensuales',
    'Total': 'Cargos_totales'
}
df.rename(columns=traduccion_columnas, inplace=True)
df.columns
```

```python
# Verifique si hay valores únicos en cada columna

for col in df.columns:
    print(f"Columna: {col}")
    print(f"Valores únicos: {df[col].nunique(dropna=False)}")
    print("Valores únicos:", df[col].unique()[:5])
    print("-" * 50)
```

```python
# Verificamos cuántas filas duplicadas existen
df.duplicated().sum()
```

```python
# Creamos una tabla resumen de nulos
resumen_nulos = pd.DataFrame({
    'Columna': df.columns,
    'Nulos': df.isnull().sum(),
    'Tipo de dato': df.dtypes
}).reset_index(drop=True)

resumen_nulos
```

```python
# Resumen de vacíos
resumen_nulos = pd.DataFrame({
    'Columna': df.columns,
    'Nulos (NaN)': df.isnull().sum().values,
    'Espacios en blanco': (df.applymap(lambda x: isinstance(x, str) and x.strip() == '')).sum().values,
    'Total vacíos': (df.isnull().sum() + df.applymap(lambda x: isinstance(x, str) and x.strip() == '')).sum().values,
})

# Mostrar solo las columnas que tienen vacíos
resumen_vacios = resumen_nulos[resumen_nulos['Total vacíos'] > 0]

resumen_vacios
```

```python
# Eliminar filas donde Abandono o Cargos_totales sean cadenas vacías ('')
df = df[~(
    (df['Abandono'].str.strip() == '') |
    (df['Cargos_totales'].str.strip() == '')
)]
df.info()
```

```python
# Crear la columna 'Cuentas_Diarias' dividiendo cargos mensuales entre 30
df['Cuentas_Diarias'] = df['Cargos_mensuales'] / 30
```

#📊 Carga y análisis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


# Configuración de estilo
sns.set(style="white")
plt.rcParams["figure.figsize"] = (10, 6)

# Configuración de colores
color_no = "#4098cf"
color_yes = "#c9521a"
```

```python
df_filtrado = df[df["Abandono"].isin(["Yes", "No"])]

# Histograma
sns.histplot(df, x='Meses_contratados',
             hue='Abandono',
             multiple='dodge',
             bins=30,
             palette={'Yes':color_yes, 'No':color_no},
             kde=False,
             alpha=0.6,)

plt.title('Comparación de Meses Contratados: Abandono vs Permanencia', fontsize=14)
plt.xlabel('Meses Contratados')
plt.ylabel('Número de Clientes')
plt.legend(title='Estado del Cliente',labels=['Abandona','Permanece'])
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
sns.kdeplot(
    data=df[df["Abandono"].isin(["Yes", "No"])],
    x='Cuentas_Diarias',
    hue='Abandono',
    fill=True,
    common_norm=False,
    palette={'Yes': color_yes, 'No': color_no},
    alpha=0.6
)
plt.title('Distribución de Densidad: Cuentas Diarias según Estado del Cliente', fontsize=14)
plt.xlabel('Cuentas Diarias')
plt.ylabel('Densidad')
plt.xlim(left=0)
plt.legend(title='Estado del Cliente', labels=['Abandona', 'Permanece'])
plt.grid(True)
plt.tight_layout()
plt.show()


```

```python
sns.boxplot(
    data=df[df["Abandono"].isin(["Yes", "No"])],
    x='Abandono',
    y='Cuentas_Diarias',
    palette={'Yes': color_yes, 'No': color_no}
)
plt.title('Cuentas Diarias por Estado del Cliente', fontsize=14)
plt.xlabel('Estado del Cliente')
plt.ylabel('Cuenta Diaria')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Gráfico de densidad (KDE)
sns.kdeplot(data=df[df['Abandono'] == 'Yes'],
            x='Meses_contratados',
            label='Abandona',
            fill=True,
            common_norm=False,
            alpha=0.5, color=color_yes, linewidth=2)

sns.kdeplot(data=df[df['Abandono'] == 'No'],
            x='Meses_contratados',
            label='Permanece',
            fill=True,
            common_norm=False, alpha=0.5, color=color_no, linewidth=2)

plt.title('Densidad de Meses Contratados: Abandono vs Permanencia', fontsize=14)
plt.xlabel('Meses Contratados')
plt.ylabel('Densidad')
plt.legend(title='Estado del Cliente')
plt.grid(True)
plt.tight_layout()
```

```python
# Creamos una copia del DataFrame filtrado
df_kde = df[df['Abandono'].isin(['Yes', 'No'])].copy()

# Codificamos las variables categóricas a números
contrato_map = {'Month-to-month': 0, 'One year': 1, 'Two year': 2}
abandono_map = {'No': 0, 'Yes': 1}

df_kde['Contrato_codificado'] = df_kde['Tipo_contrato'].map(contrato_map)
df_kde['Abandono_codificado'] = df_kde['Abandono'].map(abandono_map)

# Gráfico de densidad
sns.kdeplot(
    data=df_kde,
    x='Contrato_codificado',
    hue='Abandono',
    fill=True,
    common_norm=False,
    alpha=0.5
)

# Personalización
plt.title("Densidad de Abandono según Tipo de Contrato", fontsize=14)
plt.xlabel("Tipo de Contrato (0: Mes a mes, 1: 1 año, 2: 2 años)")
plt.ylabel("Densidad")
plt.xticks([0, 1, 2], ['Month-to-month', 'One year', 'Two year'])
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Reemplazar los valores en el DataFrame original o en una copia
df['Adulto_mayor'] = df['Adulto_mayor'].replace({0: 'No', 1: 'Sí'})
```

```python
# Filtro para valores válidos
df_filtrado = df[df["Abandono"].isin(["Yes", "No"])]

# Colores personalizados
color_no = "#4C72B0"   # Permanece
color_yes = "#DD8452"  # Abandona
palette = {'Yes': color_yes, 'No': color_no}

# Variables categóricas a comparar
columnas = ['Género', 'Adulto_mayor', 'Dependientes', 'Pareja']

# Crear figura de 2x2
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

# Guardar handles para la leyenda
handles = None

for i, col in enumerate(columnas):
    plot = sns.histplot(
        data=df_filtrado,
        x=col,
        hue='Abandono',
        multiple='dodge',
        shrink=0.8,
        palette=palette,
        kde=False,
        alpha=0.6,
        ax=axes[i]
    )
    axes[i].set_title(f'Comparación por {col.replace("_", " ")}: Abandono vs Permanencia', fontsize=14)
    axes[i].set_xlabel(col.replace("_", " "))
    axes[i].set_ylabel("Número de Clientes")
    axes[i].set_ylim(0, 4600)  # Limitar eje y
    axes[i].grid(True)

    # Solo guardar los handles y labels del primer gráfico
    if handles is None:
        handles, labels = axes[i].get_legend_handles_labels()

    # Quitar la leyenda de los subgráficos
    axes[i].get_legend().remove()

plt.show()
```

# 📊 Informe Final – Análisis de Evasión de Clientes en TelecomX

---

## 🔹 Introducción

La retención de clientes es un desafío constante en el sector de telecomunicaciones, especialmente cuando los usuarios tienen contratos flexibles y múltiples opciones en el mercado. Este análisis se enfoca en identificar los factores que influyen en la **evasión de clientes (churn)** utilizando un conjunto de datos de clientes de la empresa **TelecomX**.

El objetivo principal es comprender los perfiles de los clientes que abandonan la compañía, los servicios que más se asocian al abandono y cómo ciertos patrones de uso pueden predecir el riesgo de churn. Este conocimiento permitirá generar estrategias que mejoren la retención y optimicen la experiencia del cliente.

---

## 🔹 Limpieza y Tratamiento de Datos

Los datos fueron importados desde un archivo `.json` mediante una URL pública de GitHub. Posteriormente, se realizaron las siguientes acciones de limpieza y transformación:

- Eliminación de filas con valores nulos.
- Conversión de variables numéricas incorrectamente tipadas (`Cargos_totales`) a tipo float.
- Normalización de etiquetas (`Yes`/`No`, `No internet service`).
- Conversión de valores binarios como `Adulto_mayor` de 0/1 a `Sí`/`No`.
- Creación de nuevas variables como:
  - `Cuentas_Diarias`: costo diario del servicio (calculado a partir del cargo mensual).
  - `Servicios_Activos`: número de servicios digitales contratados por cliente.
  - `Abandono_Num`: variable binaria para análisis de correlaciones.

## 🔹 Conclusiones e Insights

- El **perfil de cliente con mayor probabilidad de abandono** es alguien con contrato *mes a mes*, menos de 10 meses en la compañía, servicios de fibra óptica, y múltiples servicios digitales (aunque esto último es paradójico, indicando frustración).
- Los clientes que **pagan más por día** tienden a abandonar con mayor frecuencia, lo cual sugiere una **sensibilidad al precio diario**, más que al mensual.
- En contraste, los clientes que **contratan más servicios activos** tienen menor tasa de abandono, lo que indica que **el valor percibido** puede fidelizar si se entrega correctamente.

---

## 🔹 Recomendaciones

Con base en los análisis realizados, se sugieren las siguientes acciones estratégicas:

1. **Fidelizar en los primeros 6 meses**: implementar programas de bienvenida, atención personalizada o incentivos para clientes en contratos flexibles.
2. **Revisión del servicio de fibra óptica**: detectar posibles fallos técnicos o expectativas no cumplidas.
3. **Optimizar la experiencia con servicios adicionales**: mejorar usabilidad, comunicación del valor agregado y soporte.
4. **Revisar el proceso de pago electrónico**: detectar fricciones o errores que generen desconfianza.
5. **Estrategias de upselling inteligentes**: promover paquetes de servicios adicionales con beneficios integrados que aumenten la retención.
6. **Ofrecer descuentos o mejoras a clientes de alto gasto diario**, ya que son más propensos a dejar la compañía si no perciben suficiente valor.

---

Este análisis proporciona una base sólida para la toma de decisiones en áreas de retención de clientes, diseño de servicios y políticas de precios. Al entender el comportamiento del cliente, **TelecomX puede anticiparse al abandono y mejorar la fidelización de su base actual**.
