#ðŸ“Œ ExtracciÃ³n

```python
import pandas as pd
```

```python
df = pd.read_json('https://raw.githubusercontent.com/francisco-gamez/Challenge-Telecom-X/refs/heads/main/TelecomX_Data.json')
df
```

```python
df_customer = df['customer'].apply(pd.Series)
df_phone = df['phone'].apply(pd.Series)
df_internet = df['internet'].apply(pd.Series)
df_account = df['account'].apply(pd.Series)
```

```python
df = pd.concat([df.drop(columns=['customer', 'phone', 'internet','account']),
                   df_customer, df_phone, df_internet,df_account], axis=1)
df
```

```python
df_Charges = df['Charges'].apply(pd.Series)
df = pd.concat([df.drop(columns=['Charges']), df_Charges], axis=1)
df
```

```python
df.info()
```

#ðŸ”§ TransformaciÃ³n

```python
# Cambiando el nombre de las columnas conforme al diccionario

traduccion_columnas = {
    'customerID': 'ID_cliente',
    'Churn': 'Abandono',
    'gender': 'GÃ©nero',
    'SeniorCitizen': 'Adulto_mayor',
    'Partner': 'Pareja',
    'Dependents': 'Dependientes',
    'tenure': 'Meses_contratados',
    'PhoneService': 'Servicio_telefÃ³nico',
    'MultipleLines': 'MÃºltiples_lÃ­neas',
    'InternetService': 'Servicio_Internet',
    'OnlineSecurity': 'Seguridad_en_lÃ­nea',
    'OnlineBackup': 'Respaldo_en_lÃ­nea',
    'DeviceProtection': 'ProtecciÃ³n_de_dispositivo',
    'TechSupport': 'Soporte_tÃ©cnico',
    'StreamingTV': 'TelevisiÃ³n_por_cable',
    'StreamingMovies': 'Streaming_de_pelÃ­culas',
    'Contract': 'Tipo_contrato',
    'PaperlessBilling': 'Factura_electrÃ³nica',
    'PaymentMethod': 'MÃ©todo_de_pago',
    'Monthly': 'Cargos_mensuales',
    'Total': 'Cargos_totales'
}
df.rename(columns=traduccion_columnas, inplace=True)
df.columns
```

```python
# Verifique si hay valores Ãºnicos en cada columna

for col in df.columns:
    print(f"Columna: {col}")
    print(f"Valores Ãºnicos: {df[col].nunique(dropna=False)}")
    print("Valores Ãºnicos:", df[col].unique()[:5])
    print("-" * 50)
```

```python
# Verificamos cuÃ¡ntas filas duplicadas existen
df.duplicated().sum()
```

```python
# Creamos una tabla resumen de nulos
resumen_nulos = pd.DataFrame({
    'Columna': df.columns,
    'Nulos': df.isnull().sum(),
    'Tipo de dato': df.dtypes
}).reset_index(drop=True)

resumen_nulos
```

```python
# Resumen de vacÃ­os
resumen_nulos = pd.DataFrame({
    'Columna': df.columns,
    'Nulos (NaN)': df.isnull().sum().values,
    'Espacios en blanco': (df.applymap(lambda x: isinstance(x, str) and x.strip() == '')).sum().values,
    'Total vacÃ­os': (df.isnull().sum() + df.applymap(lambda x: isinstance(x, str) and x.strip() == '')).sum().values,
})

# Mostrar solo las columnas que tienen vacÃ­os
resumen_vacios = resumen_nulos[resumen_nulos['Total vacÃ­os'] > 0]

resumen_vacios
```

```python
# Eliminar filas donde Abandono o Cargos_totales sean cadenas vacÃ­as ('')
df = df[~(
    (df['Abandono'].str.strip() == '') |
    (df['Cargos_totales'].str.strip() == '')
)]
df.info()
```

#ðŸ“Š Carga y anÃ¡lisis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


# ConfiguraciÃ³n de estilo
sns.set(style="white")
plt.rcParams["figure.figsize"] = (10, 6)

# ConfiguraciÃ³n de colores
color_no = "#4098cf"
color_yes = "#c9521a"
```

```python
df_filtrado = df[df["Abandono"].isin(["Yes", "No"])]

# Histograma
sns.histplot(df, x='Meses_contratados',
             hue='Abandono',
             multiple='dodge',
             bins=30,
             palette={'Yes':color_yes, 'No':color_no},
             kde=False,
             alpha=0.6,)

plt.title('ComparaciÃ³n de Meses Contratados: Abandono vs Permanencia', fontsize=14)
plt.xlabel('Meses Contratados')
plt.ylabel('NÃºmero de Clientes')
plt.legend(title='Estado del Cliente',labels=['Abandona','Permanece'])
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# GrÃ¡fico de densidad (KDE)
sns.kdeplot(data=df[df['Abandono'] == 'Yes'],
            x='Meses_contratados',
            label='Abandona',
            fill=True,
            common_norm=False,
            alpha=0.5, color=color_yes, linewidth=2)

sns.kdeplot(data=df[df['Abandono'] == 'No'],
            x='Meses_contratados',
            label='Permanece',
            fill=True,
            common_norm=False, alpha=0.5, color=color_no, linewidth=2)

plt.title('Densidad de Meses Contratados: Abandono vs Permanencia', fontsize=14)
plt.xlabel('Meses Contratados')
plt.ylabel('Densidad')
plt.legend(title='Estado del Cliente')
plt.grid(True)
plt.tight_layout()
```

```python
# Creamos una copia del DataFrame filtrado
df_kde = df[df['Abandono'].isin(['Yes', 'No'])].copy()

# Codificamos las variables categÃ³ricas a nÃºmeros
contrato_map = {'Month-to-month': 0, 'One year': 1, 'Two year': 2}
abandono_map = {'No': 0, 'Yes': 1}

df_kde['Contrato_codificado'] = df_kde['Tipo_contrato'].map(contrato_map)
df_kde['Abandono_codificado'] = df_kde['Abandono'].map(abandono_map)

# GrÃ¡fico de densidad
sns.kdeplot(
    data=df_kde,
    x='Contrato_codificado',
    hue='Abandono',
    fill=True,
    common_norm=False,
    alpha=0.5
)

# PersonalizaciÃ³n
plt.title("Densidad de Abandono segÃºn Tipo de Contrato", fontsize=14)
plt.xlabel("Tipo de Contrato (0: Mes a mes, 1: 1 aÃ±o, 2: 2 aÃ±os)")
plt.ylabel("Densidad")
plt.xticks([0, 1, 2], ['Month-to-month', 'One year', 'Two year'])
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Reemplazar los valores en el DataFrame original o en una copia
df['Adulto_mayor'] = df['Adulto_mayor'].replace({0: 'No', 1: 'SÃ­'})
```

```python
# Filtro para valores vÃ¡lidos
df_filtrado = df[df["Abandono"].isin(["Yes", "No"])]

# Colores personalizados
color_no = "#4C72B0"   # Permanece
color_yes = "#DD8452"  # Abandona
palette = {'Yes': color_yes, 'No': color_no}

# Variables categÃ³ricas a comparar
columnas = ['GÃ©nero', 'Adulto_mayor', 'Dependientes', 'Pareja']

# Crear figura de 2x2
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

# Guardar handles para la leyenda
handles = None

for i, col in enumerate(columnas):
    plot = sns.histplot(
        data=df_filtrado,
        x=col,
        hue='Abandono',
        multiple='dodge',
        shrink=0.8,
        palette=palette,
        kde=False,
        alpha=0.6,
        ax=axes[i]
    )
    axes[i].set_title(f'ComparaciÃ³n por {col.replace("_", " ")}: Abandono vs Permanencia', fontsize=14)
    axes[i].set_xlabel(col.replace("_", " "))
    axes[i].set_ylabel("NÃºmero de Clientes")
    axes[i].set_ylim(0, 4600)  # Limitar eje y
    axes[i].grid(True)

    # Solo guardar los handles y labels del primer grÃ¡fico
    if handles is None:
        handles, labels = axes[i].get_legend_handles_labels()

    # Quitar la leyenda de los subgrÃ¡ficos
    axes[i].get_legend().remove()

plt.show()
```

```python
# Histograma
sns.histplot(
    data=df_filtrado,
    x='Tipo_contrato',
    hue='Abandono',
    multiple='dodge',
    shrink=0.8,
    palette={'Yes': color_yes, 'No': color_no},
    kde=False,
    alpha=0.6,
    bins=None  # No aplica en categÃ³ricas, pero se incluye por consistencia
)

# PersonalizaciÃ³n del grÃ¡fico
plt.title('ComparaciÃ³n por Tipo de Contrato: Abandono vs Permanencia', fontsize=14)
plt.xlabel('Tipo de Contrato')
plt.ylabel('NÃºmero de Clientes')
plt.ylim(0, 2500)
plt.legend(title='Estado del Cliente', labels=['Abandona', 'Permanece'])
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Boxplot
sns.boxplot(
    data=df_filtrado,
    x='Abandono',
    y='Cargos_mensuales',
    palette=palette,
    order=['Yes', 'No']  # Para que el orden coincida con 'Abandona' / 'Permanece'
)

# PersonalizaciÃ³n
plt.title('DistribuciÃ³n de Cargos Mensuales: Abandono vs Permanencia', fontsize=14)
plt.xlabel('Estado del Cliente')
plt.ylabel('Cargos Mensuales')
plt.xticks([0, 1], ['Abandona', 'Permanece'])
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Asegurarse de que los cargos totales sean numÃ©ricos
df['Cargos_totales'] = pd.to_numeric(df['Cargos_totales'], errors='coerce')

# Filtrar solo valores vÃ¡lidos para graficar
df_filtrado = df[(df['Cargos_totales'] >= 0) & (df['Abandono'].isin(['Yes', 'No']))].copy()

# GrÃ¡fico de densidad
sns.kdeplot(
    data=df_filtrado,
    x='Cargos_totales',
    hue='Abandono',
    fill=True,
    common_norm=False,
    palette={'Yes': color_yes, 'No': color_no},
    alpha=0.6
)

# PersonalizaciÃ³n del grÃ¡fico
plt.title('DistribuciÃ³n de Densidad: Cargos Totales segÃºn Estado del Cliente', fontsize=14)
plt.xlabel('Cargos Totales')
plt.ylabel('Densidad')
plt.legend(title='Estado del Cliente', labels=['Abandona', 'Permanece'])
plt.grid(True)
plt.xlim(left=0)  # Limitar eje X a no mostrar negativos
plt.tight_layout()
plt.show()

```
